```python
# using selenium to fetch product names, price and links of computers
from selenium import webdriver
```

```python
url = 'https://www.jumia.com.gh/computing-accessories/'
```

```python
driver = webdriver.Chrome()
```

```python
# Set an implicit wait
driver.implicitly_wait(10)
driver.get(url) # launches a web browser and get url
```

```python
# find product name by XPath
product = [name.text for name in driver.find_elements_by_xpath("//h3")]
len(product)
product
```

```python
# find product price by XPath
price = [price.text for price in driver.find_elements_by_xpath('//div[@class= "info"]//child::div[@class="prc"]')]
len(price)
price
```

```python
# find links by XPath
link = [url.get_attribute("href") for url in driver.find_elements_by_xpath('//article[contains(@class,"prd _fb")]//child::a')]
len(link)
link
```

```python
import pandas as pd
# import pandas library
Laptops = pd.DataFrame({
    "Product": product,
    "Price": price,
    "Link":link 
})
Laptops
# put results into a dataframe
```

```python
Laptops.to_csv("jumia_computers.csv")
# create a csv file and send results to it
```
